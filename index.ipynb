# Rename all the columns to lower case

```python
#Your code here
lower_cols = []
for col in df.columns:
    lower_cols.append(col.lower())
df.columns = lower_cols
```

## Compare subway traffic by day of the week. Display this as a graph.

```python
#Your code here
df.DATE = pd.to_datetime(df.DATE)
```

```python
df['DAYOFWEEK'] = df.DATE.dt.dayofweek
```

## Is there more subway traffic on a weekend or a weekday?    Be specific in comparing magnitudes.

```python
gb = gb.reset_index()
gb.columns
```

```python
#Your code here
#create weekend bool column using .map()
gb['ISWEEKEND'] = gb.DAYOFWEEK.map({0:False,1:False,2:False,3:False,4:False,5:True,6:True})
print(gb)
```

```python
#Getting a mean for 
wknd = gb.groupby('ISWEEKEND').mean()
print(wknd)
```

```python
#Selecting entries/exits for plotting horzintally
wknd[['ENTRIES', 'EXITS']].plot(kind='barh')
```
